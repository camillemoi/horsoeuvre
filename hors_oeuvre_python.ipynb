```python
import folium
import numpy as np
```

```python
x_med = np.median(df['latitude'])
y_med = np.median(df['longitude'])
m = folium.Map([x_med, y_med], zoom_start=14, control_scale=True)
```

```python
m
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster()
```

```python
def color(categ):
    if categ == 'Alimentaire':
        col = 'blue'
    elif categ == 'Service':
        col = 'red'
    elif categ == 'Bar Café Restaurant':
        col = 'beige'
    elif categ == 'Loisirs Sport Culture':
        col = 'purple'
    elif categ == 'Non Alimentaire':
        col = 'pink'
    else:
        col = 'darkgreen'
    return col
```

```python
def logo(categ):
    if categ == 'Alimentaire':
        log = 'cart-plus'
    elif categ == 'Service':
        log = 'phone'
    elif categ == 'Bar Café Restaurant':
        log = 'cutlery'
    elif categ == 'Loisirs Sport Culture':
        log = 'book'
    elif categ == 'Non Alimentaire':
        log = 'shopping-bag'
    else:
        log = 'question-circle'
    return log
```

```python
for lat,lan,name,categ in zip(df['latitude'],df['longitude'],df['FONCTION'],df['TYPE']):
    mc.add_child(folium.Marker(location=[lat,lan],popup = name,
                  icon= folium.Icon(color=color(categ),
                  icon_color='black',icon =logo(categ), prefix='fa')))
map = m.add_child(mc)
map
```

```python
#export HTML
outfp = "carte_finale.html"
m.save(outfp)
```
